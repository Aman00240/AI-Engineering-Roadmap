```python
from typing import Iterable
from dataclasses import dataclass

@dataclass
class User:
    name:str
    age:int

def process_data(items:Iterable[int])-> list[int]:
    return [i*2 for i in items if i>10]

def get_user_info(user:User)-> str:
    return f"User:{user.name} , Age: {user.age}"

data = [5, 12, 3, 20, 8]
print(process_data(data))
user_data = User(name="Alice",age=30)
print(get_user_info(user_data))
```

```python
# bad_code.py 
data = [
    {"name": "Alice", "age": 30, "score": 85.5},
    {"name": "Bob", "age": "twenty", "score": 90},
    {"name": "Charlie", "age": 25, "score": None},
]

def process_users(users):
    valid_users = []
    for user in users:
        if isinstance(user['age'], int) and user['score'] is not None:
            valid_users.append(user)
    return valid_users

def get_average_score(users):
    if not users:
        return 0
    total = 0
    for user in users:
        total += user['score']
    return total / len(users)

cleaned = process_users(data)
avg = get_average_score(cleaned)
print(f"Average score: {avg}")
```

```python
from dataclasses import dataclass

data: list[dict[str, str | int | float | None]] = [
    {"name": "Alice", "age": 30, "score": 85.5},
    {"name": "Bob", "age": "twenty", "score": 90},
    {"name": "Charlie", "age": 25, "score": None},
]


@dataclass
class RawUser:
    name: str
    age: int | str
    score: int | float | None  
    
@dataclass
class CleanUser:
    name: str
    age: int 
    score: int | float  

def process_users(raw_data:list[dict])-> list[CleanUser]:
    valid_users = []
    
    for item  in raw_data:
        current_raw=RawUser(**item)
        
        if isinstance(current_raw.age,int) and current_raw.score is not None:
            clean=CleanUser(
                name=current_raw.name, 
                age=current_raw.age, 
                score=current_raw.score
            )
            valid_users.append(clean)

    return valid_users
    
def get_average_score(users:list[CleanUser])-> float:
    return sum(user.score for user in users)/len(users)
    
cleaned = process_users(data)
avg = get_average_score(cleaned)
print(f"Average score: {avg}")
```
