```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
print("LOding Embedding model")
model = SentenceTransformer("all-MiniLM-L6-v2")

sentences = ["This is a happy cat", "That is a joyous kitty", "It is raining outside"]

print("Converting text to vectors")
embeddings = model.encode(sentences)


print(f"Vector Shape:{embeddings.shape}")
print(f"First 5 numbers of 'cat':{embeddings[0][:5]}")


score_cat_kitty = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0] #type:ignore
score_cat_rain = cosine_similarity([embeddings[0]], [embeddings[2]])[0][0] #type:ignore

print("RESULTS")
print(f"Similarity (Cat vs Kitty): {score_cat_kitty:.4f}")
print(f"Similarity (Cat vs Rain):  {score_cat_rain:.4f}")


if score_cat_kitty > 0.5:
    print("\nSUCCESS: The AI understands they are related!")
else:
    print("\nFAIL: The AI is confused.")
```
