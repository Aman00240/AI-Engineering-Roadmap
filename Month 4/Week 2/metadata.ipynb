```python
import chromadb
```

```python
client=chromadb.PersistentClient(path="./chroma_db")
```

```python
try:
    client.delete_collection(name="tagged_docs")
except:
    pass
```

```python
collection=client.create_collection(name="tagged_docs")
```

```python
collection.add(
    documents=[
        "Meeting with the boss at 5 PM.",      
        "Dinner party with friends at 5 PM."   
    ],
    ids=["doc1","doc2"],
    metadatas=[
        {"category":"work"},
        {"category":"personal"}
    ]
)
```

```python
query="What is happening at 5 PM?"

results_no_filter=collection.query(
    query_texts=[query],
    n_results=2
)
```

```python
for doc,meta in zip(results_no_filter['documents'][0],results_no_filter['metadatas'][0]):
    print(f"found:{doc}, (Tag:{meta})")
```

```python
results_no_filter=collection.query(
    query_texts=[query],
    n_results=2,
    where={"category":"work"}
)
```

```python
for doc,meta in zip(results_no_filter['documents'][0],results_no_filter['metadatas'][0]):
    print(f"found:{doc}, (Tag:{meta})")
```
