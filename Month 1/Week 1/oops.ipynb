```python
class StudentLagacy:
    def __init__(self,name:str,age:int,grades:list[int]):
        self.name=name
        self.age=age
        self.grades=grades
    
    def get_average(self)->float:
        
        if not self.grades:
            return 0.0
        
        total=0
        for i in self.grades:
            total+=i
        
        return total/len(self.grades)
        
stu1=StudentLagacy('Ram',13,[78,65,67])

print(f"Average:{stu1.get_average()}")
```

```python
from dataclasses import dataclass,field

@dataclass
class StudentLagacy:
    name:str
    age:int
    grades:list[int]=field(default_factory=list)
    
    def get_average(self)->float:
        
        if not self.grades:
            return 0.0
        
        total=0
        for i in self.grades:
            total+=i
        
        return total/len(self.grades)
    
stu1=StudentLagacy('Rem',17,[88,77,66])

print(f"Average:{stu1.get_average()}")
```
