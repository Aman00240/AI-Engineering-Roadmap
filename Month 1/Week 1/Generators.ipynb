```python
import random 
from typing import Iterator
import time

def stream_logs()->Iterator[str]:
    log_db = [
        "INFO: MISTER YOU ARE HACKED",
        "INFO: YES YOU ARE",
        "ERROR: Uuuh did you just do rm -rf /?",
        "INFO: clicked WRONG button",
        "ERROR: 404 Brain not Found",
        "ERROR: Payment declined(It is ok I know you are broke)"
    ]
    while True:
        random_string=random.choice(log_db)
        yield random_string
        
        

def filter_errors(source_generator:Iterator[str])->Iterator[str]:
        for line in source_generator:
            if "ERROR" in line:
                yield line
        

def batch_data(source:Iterator[str],batch_size:int)->Iterator[list[str]]:
    bucket:list[str]=[]
    
    for error in source:
        bucket.append(error)
        
        if len(bucket)==batch_size:
            yield bucket
            bucket=[]
            
my_stream=stream_logs()

error_logs=filter_errors(my_stream)

data=batch_data(error_logs,3)

print(next(data))
print(next(data))
print(next(data))

```
