```python
import chromadb
```

```python
client=chromadb.PersistentClient(path="./chroma_db")
```

```python
collection=client.get_or_create_collection(name="my_notes")
```

```python
collection.upsert(
    documents=[
        "I need to buy groceries: milk, eggs, and bread.",
        "The project deadline is next Friday.",
        "To fix the bug, restart the server and clear cache.",
        "My favorite movie is Inception."
    ],
    ids=["note1","note2","note3","note4"]
)
```

```python
user_query="What do i need to get from store?"

results=collection.query(
    query_texts=[user_query],
    n_results=1 # top 1st best match 
)
results
```

```python
print(f"Document: {results['documents'][0][0]}")
print(f"Distance: {results['distances'][0][0]:.4f}")
```
