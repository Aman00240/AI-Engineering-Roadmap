```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import numpy as np
```

```python
model=SentenceTransformer('all-MiniLM-L6-v2')
```

```python
documents = [
    "Python is a versatile programming language widely used in AI.",  
    "The best way to cook pasta is to salt the boiling water.",       
    "Machine learning models require large datasets for training.",   
    "A good sourdough bread takes 24 hours to ferment.",              
    "FastAPI is a modern web framework for building APIs."            
]
```

```python
query="Tell me about coding for artificial intelligence"

doc_embeddings=model.encode(documents)
query_embeddings=model.encode([query])

scores=cosine_similarity(query_embeddings,doc_embeddings)[0]

best_doc_index=np.argmax(scores)
best_score=scores[best_doc_index]


print("SEARCH RESULT")
print(f"Query: '{query}'")
print(f"Best Match (Index {best_doc_index}):")
print(f"Content: \"{documents[best_doc_index]}\"")
print(f"Score: {best_score:.4f}")
```
